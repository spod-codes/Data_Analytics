```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\spodc\Downloads\airlines_reviews.csv\airlines_reviews.csv")

# Basic info
df.info()

# Preview
df.head()


# Rename columns (remove spaces, lowercase)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert 'recommended' to binary (1 = yes, 0 = no)
df['recommended'] = df['recommended'].str.lower().map({'yes': 1, 'no': 0})

# Check missing values
print("Missing values:\n", df.isnull().sum())

# drop rows with missing important values
df.dropna(subset=['overall_rating', 'recommended'], inplace=True)


# Overall summary
df.describe()

# Average overall rating
avg_rating = df['overall_rating'].mean()
print(f"Average overall rating: {round(avg_rating, 2)}")

# % of people who recommended
recommend_rate = df['recommended'].mean() * 100
print(f"Percentage of people who recommended: {round(recommend_rate, 2)}%")

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='class', hue='recommended', data=df)
plt.title('Recommendations by Travel Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Recommended')
plt.xticks(rotation=45)
plt.show()

sns.boxplot(x='type_of_traveller', y='overall_rating', data=df)
plt.title('Overall Rating by Traveller Type')
plt.xticks(rotation=45)
plt.show()


numerical_cols = [
    'seat_comfort', 'staff_service', 'food_&_beverages',
    'inflight_entertainment', 'value_for_money', 'overall_rating'
]

corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title("Correlation with Overall Rating")
plt.show()
```
